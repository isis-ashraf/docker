```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('books.csv')
data
```

```python
data.info()
```

```python
# columns with missing values
missing_values = data.columns[data.isnull().sum() > 0]
missing_values
```

```python
# handling columns with missing values

# dropping data with small impact on analysis 
data.drop(['isbn', 'isbn13'], axis=1, inplace=True)

# replacing 'original-title' with values in 'title' column
data['original_title'].fillna(data['title'], inplace=True)

# replacing (language-code, original-publication-year) with placeholder
data['language_code'].fillna('unknown', inplace=True)
data['original_publication_year'].fillna(0, inplace=True)

# dropping image columns as they aren't necessary for my analysis
data.drop(['image_url', 'small_image_url'], axis=1, inplace=True)

# data after preprocessing
data
```

```python
# focusing on Harry Potter books
harry_potter_books = data[data['original_title'].str.contains('Harry Potter')]
harry_potter_books
```

```python
# most selling 5 books of Harry Potter series
most_selling = harry_potter_books.sort_values(by='ratings_count', ascending=False)
most_selling.head()
```

```python
# average rating of Harry Potter books
average_rating = harry_potter_books['average_rating'].mean()
print("Average rating of Harry Potter books:", average_rating)
```
